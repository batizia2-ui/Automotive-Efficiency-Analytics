```python
import pandas as pd

# Load manufacturing data
data = {
    'Line': ['Line A'],
    'Planned_Time_Min': [480],
    'Down_Time_Min': [70],
    'Total_Units': [3000],
    'Scrap_Units': [120],
    'Ideal_Cycle_Time_Sec': [8]
}

df = pd.DataFrame(data)

# KPI Calculations
df['Availability'] = (df['Planned_Time_Min'] - df['Down_Time_Min']) / df['Planned_Time_Min']
df['Performance'] = (df['Total_Units'] * df['Ideal_Cycle_Time_Sec']) / ((df['Planned_Time_Min'] - df['Down_Time_Min']) * 60)
df['Quality'] = (df['Total_Units'] - df['Scrap_Units']) / df['Total_Units']
df['OEE'] = df['Availability'] * df['Performance'] * df['Quality']

# Display results in English for the Portfolio
print("--- Manufacturing Efficiency Results ---")
print(df[['Line', 'Availability', 'Performance', 'Quality', 'OEE']])
```
